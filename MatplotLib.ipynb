```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
x2 = [1,2,3,4,5]

y = [1,2,3,4,5]
y2 = [5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2,"y--")
plt.show()
```

```python
x = np.array([1,2,3,4,5,6])
y = 2*(x**4) + 100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = 2*(x**4) + 100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4
y2 = x**3

plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2, color="blue", label="x^2")
plt.legend()
plt.grid()
plt.title("Example Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([0,15,0,20000])
plt.show()
```

```python
profit = [10,20,30,25]
labels = ["flipkart", "amazon", "jiomart", "myntra"]

colors = ["red", "yellow", "pink", "blue"]
explode = [0.1, 0.05, 0, 0]

plt.pie(profit, labels=labels, colors=colors, autopct="%.2f", explode=explode)
plt.show()
```
